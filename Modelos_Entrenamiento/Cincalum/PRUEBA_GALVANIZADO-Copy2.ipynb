```python
data=pd.read_excel('Cert_Cincalum_19_01_21.xlsx')
dt=joblib.load('ModeloCincalum.pkl')
enc=joblib.load('TargetCincalum.pkl')
encobj=joblib.load('EncodificacionObjetivo.pkl')
dtc=DecisionTreeClassifier()
data
```

```python
data=pd.DataFrame(data)
data
```

```python
databuena=data.loc[:, ['CCLASEPRODTOOrigen', 'ACEROOrigen', 'ACABSUPERFRECOrigen',
       'ACEITADOOrigen', 'BORDESOrigen', 'CARAINSPECCIONOrigen',
       'DIAM__INTERNOOrigen', 'EXTREMOSOrigen', 'NORMAOrigen',
       'NORRECUBRIMIENOrigen', 'NORTOLANCHOOrigen', 'NORTOLESPESOROrigen',
       'PLANITUDOrigen', 'RECUBRIMIENTOOrigen', 'SELLOOrigen',
       'SUPERFICIEOrigen', 'TAMFLOROrigen', 'TRATSUPERFOrigen',
       'USOGRALOrigen', 'CCLASEPRODTODestino', 'ACABSUPERFRECDestino',
       'ACERODestino', 'ACEITADODestino', 'BORDESDestino',
       'CARAINSPECCIONDestino', 'DIAM__INTERNODestino', 'EXTREMOSDestino',
       'NORMADestino', 'NORRECUBRIMIENDestino', 'NORTOLANCHODestino',
       'NORTOLESPESORDestino', 'PLANITUDDestino', 'RECUBRIMIENTODestino',
       'SELLODestino', 'SUPERFICIEDestino', 'TAMFLORDestino',
       'TRATSUPERFDestino', 'USOGRALDestino']]
databuena
```

```python
databuena.columns
```

```python
datae=enc.transform(databuena)
datae
```

```python
predic=dt.predict(datae)
```

```python
Resultados=pd.DataFrame(data=predic,columns=['Reasignacion'])
Resultados.head
```

```python
ResultadosF=encobj.inverse_transform(Resultados)
ResultadosF=pd.DataFrame(data=ResultadosF,columns=['Reasignacion'])
ResultadosF
```

```python
concatenated = pd.concat([data, ResultadosF],axis=1) 
concatenated
```

```python
datae_train,datae_test,Resultados_train,Resultados_test=train_test_split(datae,Resultados)
```

```python
dtc.fit(datae_train,Resultados_train)
```

```python
dtc.score(datae_train,Resultados_train)
```

```python
dtc.score(datae_test,Resultados_test)
```

```python
# Separando los datos en sets de entrenamiento y evaluación
datae_train, datae_test, Resultados_train, Resultados_test = train_test_split(datae, 
                        Resultados, random_state=1)


tree = DecisionTreeClassifier(max_depth=2, random_state=0)
tree.fit(datae_train, Resultados_train)
print('Precisión modelo inicial train/test  {0:.3f}/{1:.3f}'
      .format(tree.score(datae_train, Resultados_train), tree.score(datae_test, Resultados_test)))
```

```python
# Utilizando AdaBoost para aumentar la precisión
ada = AdaBoostClassifier(base_estimator=tree, n_estimators=500, 
                         learning_rate=1.5, random_state=1)
# Ajustando los datos
ada = ada.fit(datae_train, Resultados_train)
```

```python
# Imprimir la precisión.
y_train_pred = ada.predict(datae_train)
y_test_pred = ada.predict(datae_test)
ada_train = accuracy_score(Resultados_train, y_train_pred)
ada_test = accuracy_score(Resultados_test, y_test_pred)
print('Precisión modelo con AdaBoost train/test  {0:.3f}/{1:.3f}'
      .format(ada_train, ada_test))
```

```python
# Armando el modelo con parametro max_depth
gbrt = GradientBoostingClassifier(random_state=0, n_estimators=500,
                                  max_depth=1, learning_rate=0.01)
# Ajustando el modelo
gbrt.fit(datae_train, Resultados_train)
print('Precisión Gradient Boosting train/test  {0:.3f}/{1:.3f}'
      .format(gbrt.score(datae_train, Resultados_train), gbrt.score(datae_test, Resultados_test)))
```
