```python
data=pd.read_excel("Cert_Prepintado_19_01_21.xlsx")
dt=joblib.load('ModeloPrepintado.pkl')
enc=joblib.load('TargetPrepintado.pkl')
encobj=joblib.load('EncodificacionObjetivo.pkl')
data
```

```python
data=pd.DataFrame(data)
data
```

```python
databuena=data.loc[:, ['CCLASEPRODTOOrigen', 'CARAINSPECCIONOrigen',
       'COLORCARA_INF_Origen', 'COLORCARA_SUP_Origen', 'DIAM__INTERNOOrigen',
       'ESQUEMAPINTADOOrigen', 'EXTREMOSOrigen', 'NORMAOrigen',
       'NORPREPINTADOOrigen', 'NORTOLANCHOOrigen', 'NORTOLESPESOROrigen',
       'PINT_C_INF_Origen', 'PINT_C_SUP_Origen', 'PLANITUDOrigen',
       'SELLOOrigen', 'SUPERFICIEOrigen', 'SUSTRATOOrigen',
       'SUSTRATORECUBROrigen', 'SUSTTRATC_INFOrigen', 'SUSTTRATC_SUPOrigen',
       'TIPOPOLIETOrigen', 'USOGRALOrigen', 'CCLASEPRODTODestino',
       'CARAINSPECCIONDestino', 'COLORCARA_INF_Destino',
       'COLORCARA_SUP_Destino', 'DIAM__INTERNODestino',
       'ESQUEMAPINTADODestino', 'EXTREMOSDestino', 'NORMADestino',
       'NORPREPINTADODestino', 'NORTOLANCHODestino', 'NORTOLESPESORDestino',
       'PINT_C_INF_Destino', 'PINT_C_SUP_Destino', 'PLANITUDDestino',
       'SELLODestino', 'SUPERFICIEDestino', 'SUSTRATORECUBRDestino',
       'SUSTRATODestino', 'SUSTTRATC_INFDestino', 'SUSTTRATC_SUPDestino',
       'TIPOPOLIETDestino', 'USOGRALDestino']] 
databuena
```

```python
databuena.columns
```

```python
datae=enc.transform(databuena)
datae
```

```python
predicprob=dt.predict_proba(datae)
predic=dt.predict(datae)
```

```python
ResultadosProb=pd.DataFrame(data=predicprob,columns=['Prob. Ok','Prob. Rechazado'])
ResultadosNormal=pd.DataFrame(data=predic,columns=['Reasignación'])
ResultadosNormal=pd.DataFrame(data=encobj.inverse_transform(ResultadosNormal),columns=['Reasignacion'])
ResultadosNormal
```

```python
concatenated = pd.concat([data, ResultadosNormal,ResultadosProb],axis=1) 
concatenated
```
