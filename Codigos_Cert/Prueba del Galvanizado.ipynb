```python
data=pd.read_excel("Cert_Galvanizado_19_01_21.xlsx")
dt=joblib.load('ModeloGalvanizado.pkl')
enc=joblib.load('TargetGalvanizado.pkl')
encobj=joblib.load('EncodificacionObjetivo.pkl')
data
```

```python
data=pd.DataFrame(data)
data
```

```python
databuena=data.loc[:, ['CCLASEPRODTOOrigen', 'ACABSUPERFRECOrigen', 'ACEITADOOrigen',
       'BORDESOrigen', 'CARAINSPECCIONOrigen', 'DIAM__INTERNOOrigen',
       'EXTREMOSOrigen', 'NORMAOrigen', 'NORRECUBRIMIENOrigen',
       'NORTOLANCHOOrigen', 'NORTOLESPESOROrigen', 'PLANITUDOrigen',
       'RECUBRIMIENTOOrigen', 'SELLOOrigen', 'SUPERFICIEOrigen',
       'TAMFLOROrigen', 'TRATSUPERFOrigen', 'USOGRALOrigen',
       'CCLASEPRODTODestino', 'ACABSUPERFRECDestino', 'ACEITADODestino',
       'BORDESDestino', 'CARAINSPECCIONDestino', 'DIAM__INTERNODestino',
       'EXTREMOSDestino', 'NORMADestino', 'NORRECUBRIMIENDestino',
       'NORTOLANCHODestino', 'NORTOLESPESORDestino', 'PLANITUDDestino',
       'RECUBRIMIENTODestino', 'SELLODestino', 'SUPERFICIEDestino',
       'TAMFLORDestino', 'TRATSUPERFDestino', 'USOGRALDestino']] 
databuena
```

```python
databuena.columns
```

```python
datae=enc.transform(databuena)
datae
```

```python
predicprob=dt.predict_proba(datae)
predic=dt.predict(datae)
```

```python
ResultadosProb=pd.DataFrame(data=predicprob,columns=['Prob. Ok','Prob. Rechazado'])
ResultadosNormal=pd.DataFrame(data=predic,columns=['Reasignación'])
ResultadosNormal=pd.DataFrame(data=encobj.inverse_transform(ResultadosNormal),columns=['Reasignacion'])
ResultadosNormal
```

```python
concatenated = pd.concat([data, ResultadosNormal,ResultadosProb],axis=1) 
concatenated
```

```python
concatenated.to_excel("ResultadoGalvanizado2021Final.xlsx")  
```
