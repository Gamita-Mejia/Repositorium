```python
data=pd.read_csv("CertCincalum.csv")
dt=joblib.load('ModeloCincalum.pkl')
enc=joblib.load('TargetCincalum.pkl')
encobj=joblib.load('EncodificacionObjetivo.pkl')
data
```

```python
data=pd.DataFrame(data)
data
```

```python
databuena=data.loc[:, ['ACABSUPERFRECDestino', 'ACABSUPERFRECOrigen', 'ACEITADODestino',
       'ACEITADOOrigen', 'BORDESDestino', 'BORDESOrigen',
       'CARAINSPECCIONDestino', 'CARAINSPECCIONOrigen', 'CCLASEPRODTODestino',
       'CCLASEPRODTOOrigen', 'DIAM__INTERNODestino', 'DIAM__INTERNOOrigen',
       'EXTREMOSDestino', 'EXTREMOSOrigen', 'NORMADestino', 'NORMAOrigen',
       'NORRECUBRIMIENOrigen', 'NORRECUBRIMIENDestino', 'NORTOLANCHODestino',
       'NORTOLANCHOOrigen', 'NORTOLESPESORDestino', 'NORTOLESPESOROrigen',
       'PLANITUDDestino', 'PLANITUDOrigen', 'RECUBRIMIENTODestino',
       'RECUBRIMIENTOOrigen', 'SELLODestino', 'SELLOOrigen',
       'SUPERFICIEDestino', 'SUPERFICIEOrigen', 'TAMFLORDestino',
       'TAMFLOROrigen', 'TRATSUPERFDestino', 'TRATSUPERFOrigen',
       'USOGRALDestino', 'USOGRALOrigen']] 
databuena
```

```python
databuena.columns
```

```python
datae=enc.transform(databuena)
datae
```

```python
predicprob=dt.predict_proba(datae)
predic=dt.predict(datae)
```

```python
ResultadosProb=pd.DataFrame(data=predicprob,columns=['Prob. Ok','Prob. Rechazado'])
ResultadosNormal=pd.DataFrame(data=predic,columns=['Reasignación'])
ResultadosNormal=pd.DataFrame(data=encobj.inverse_transform(ResultadosNormal),columns=['Reasignacion'])
ResultadosNormal
```

```python
concatenated = pd.concat([data, ResultadosNormal, ResultadosProb],axis=1) 
concatenated
```
